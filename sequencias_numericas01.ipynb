```python
import matplotlib.pyplot as plt


def a(n):        
    return (n/(n+1))


option = int(input(" 1 para sequencia convergente \n 2 para sequencia limitada e divergente \n 3 para sequencia nao limitada \n 4 para sequencia de limitacao e convergencia indefinidas \n\nDigite sua opção: "))
    
    
if( option == 1):
    print("\nSequencia convergente")
    L = float(input("\nDigite o limite: "))
    E = float(input("\nDigite a tolerancia: "))
    nmax = int(input("\nDigite o inteiro positivo Nmax: "))
    nmin = int(input("\nDigite o inteiro positivo Nmin menor que Nmax: "))
    y1 = L + E
    y2 = L
    y3 = L - E
    print("\n")
    
    print(" Tabela\n")
    print("n |   a(n)")
    
    for n in range(nmin,nmax):
        print(n,"|  ",a(n))
        plt.scatter(n,a(n))
        
    print("\n\n      Gráfico")
    
    plt.axhline(y=nmax, color='r', linestyle='-')
    plt.axhline(y=nmin, color='r', linestyle='-')
    plt.axhline(y1, color='b', linestyle='-', label='y = L + E')
    plt.axhline(y2, color='g', linestyle='-', label='y = L ')
    plt.axhline(y3, color='y', linestyle='-', label='y = L - E ')
    plt.legend()
    plt.show()

   
elif( option == 2):
    print("\nSequencia limitada e divergente")
    nmax = int(input("\nDigite o inteiro positivo Nmax: "))
    nmin = int(input("\Digite o inteiro positivo Nmin menor que Nmax: "))
    Mmax = int(input("\nDigite o valor da constante Mmax: "))
    Mmin = int(input("\nDigite o valor da constante Mmin: "))
    y1 = Mmin
    y2 = Mmax
    print("\n")
    
    print(" Tabela\n")
    print("n |   a(n)")
    
    for n in range(nmin,nmax):
        print(n,"|  ",a(n))
        plt.scatter(n,a(n))
        
    print("\n\n      Gráfico")
    
    plt.axhline(y=nmax, color='r', linestyle='-')
    plt.axhline(y=nmin, color='r', linestyle='-')
    plt.axhline(y1, color='b', linestyle='-', label='y = Mmin')
    plt.axhline(y2, color='g', linestyle='-', label='y = Mmax ')
    plt.legend()
    plt.show()

elif( option == 3):
    print("\nSequencia nao limitada")
    nmax = int(input("\nDigite o inteiro positivo Nmax: "))
    nmin = int(input("\nDigite o inteiro positivo Nmin menor que Nmax: "))
    M = float(input("\nDigite a constante M: ")) 
    y1 = M
    print("\n")
    
    print(" Tabela\n")
    print("n |   a(n)")
    
    for n in range(nmin,nmax):
        print(n,"|  ",a(n))
        plt.scatter(n,a(n))
        
    print("\n\n      Gráfico")
    
    plt.axhline(y=nmax, color='r', linestyle='-')
    plt.axhline(y=nmin, color='r', linestyle='-')
    plt.axhline(y1, color='b', linestyle='-', label='y = M')
    plt.legend()
    plt.show()
    
elif( option == 4):
    print("\nSequencia de limitacao e convergencia indefinidas")
    nmax = int(input("\nDigite o inteiro positivo Nmax: "))
    nmin = int(input("\nDigite o inteiro positivo Nmin menor que Nmax: "))
    print("\n")
    
    print(" Tabela\n")
    print("n |   a(n)")
    
    for n in range(nmin,nmax):
        print(n,"|  ",a(n))
        plt.scatter(n,a(n))
        
    print("\n\n      Gráfico")
    
    plt.axhline(y=nmax, color='r', linestyle='-')
    plt.axhline(y=nmin, color='r', linestyle='-')
    plt.show()

    
```
